```python
import matplotlib.pyplot as plt

x = [a for a in range(int(input('Pick your size of array: ')))]
y = []

for num in x:
    y.append(num**2)

match input('Choose your type of graph (plot, bar, scatter): '):
    case 'plot':
        plt.plot(x, y)
    case 'bar':
        plt.bar(x, y)
    case 'scatter':
        plt.scatter(x, y, linewidths=0)
    case _:
        print('Invalid input. Please choose from plot, bar, or scatter.')

plt.show()
plt.close()
```
